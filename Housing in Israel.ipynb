```python
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.io import push_notebook
from bokeh.charts import Chart

def parse_data(desc, data):
    desc.columns = ["id", "value"]        
    desc.columns = ['id', 'value']
    desc = desc.set_index('id')['value'].to_dict()
    data.columns = ["Date", "Value"]

    data['Date'] = pd.to_datetime(data["Date"], format="%d/%m/%Y")
    return desc,data
    
def get_html_data(filename):
    return parse_data(*pd.read_html("data\\" + filename))


```

```python
# Housing starts vs Housing completions

# data's base is 100 at 2005
sfl = get_html_data("SFL_MADAD.M.xls")
ffl = get_html_data("FFL_MADAD.M.xls")
output_notebook()
ch = figure(title="Housing Starts & Completions -- Figures are precentages (compared to base 2005)", plot_height=400, plot_width=600, x_axis_type="datetime")
ch.line(sfl[1]['Date'],sfl[1]['Value'], legend=sfl[0]['Description'])
ch.line(ffl[1]['Date'],ffl[1]['Value'], legend=ffl[0]['Description'], color='red', line_dash=[2,2], line_width=3)

deficit = pd.Series(sfl[1]['Value'] - ffl[1]['Value'])
df = figure(title="Construction Deficit | (Starts - Completions)", x_axis_type="datetime")
df.line(sfl[1]['Date'], deficit, legend="Deficit (Positive == More starts than Completions)")
push_notebook()
show(ch,notebook_handle=True)
show(df)
```

The 1995-2000 Era is quite interesting (More projects getting finished every year by far after a few years of building scaling up, then winding down), I haven't heard a good explanation why Israel in 2017 can't build at 1/3 the rate it did in 1995

There is clearly nothing on the horizon in the market today in housing starts, the start numbers are nowhere near the requirements by the market, This is a sellers market and it will keep going that way all things staying the same (Pop. expansion, Housing starts, GDP)
